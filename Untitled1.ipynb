## Importing Libraries and loading data

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("dataset/train.csv")
```

```python
data.head()
```

## Checking missing values

```python
missing = data.isnull().sum()
missing = missing[missing>0]
missing.sort_values(inplace=True)
plt.figure(figsize=(15,8))
missing.plot.bar()
```

## Plotting distribution of Sales price

```python
sns.set(rc={'figure.figsize':(12,8)})
sns.distplot(data['SalePrice'],bins = 20)
```

```python
data["SalePrice"].describe()
```

## Selecting numerical features

```python
num_features = data.select_dtypes(include = np.number)
sns.heatmap(num_features.corr())
print(num_features.corr().sort_values('SalePrice')["SalePrice"])
```

### Selecting Top 11 correlated numerical features

```python
k=11
cols = num_features.corr().nlargest(k,'SalePrice')["SalePrice"].index
sns.heatmap(data[cols].corr())
```

## Scatter plot for different variables

```python
_, ((ax1,ax2),(ax3,ax4),(ax5,ax6)) = plt.subplots(nrows=3,ncols=2,figsize=(14,10))
sns.regplot(x='OverallQual',y='SalePrice',data = data,fit_reg=True,scatter=True,ax=ax1)
sns.regplot(x='GrLivArea',y='SalePrice',data = data,fit_reg=True,scatter=True,ax=ax2)
sns.regplot(x='GarageArea',y='SalePrice',data = data,fit_reg=True,scatter=True,ax=ax3)
sns.regplot(x='FullBath',y='SalePrice',data = data,fit_reg=True,scatter=True,ax=ax4)
sns.regplot(x='YearBuilt',y='SalePrice',data = data,fit_reg=True,scatter=True,ax=ax5)
sns.regplot(x='WoodDeckSF',y='SalePrice',data = data,fit_reg=True,scatter=True,ax=ax6)
```
